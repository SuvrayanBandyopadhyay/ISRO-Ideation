```python
#Take two images
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import zoom


#Import the images
img1 = Image.open("img1.png")
img2 = Image.open("img2.png")

arr1 = np.array(img1)
arr2 = np.array(img2)

val1 = np.array(arr1[:,:,2])
val2 = np.array(arr2[:,:,0])



```

```python
#Display grayscale of both maps

# Visualize the blue channel
plt.imshow(val1, cmap='gray')
plt.title("Element 1")

plt.show()
```

```python
# Visualize the blue channel
plt.imshow(val2, cmap='gray')
plt.title("Element 2")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Create an empty RGB image
height, width = val1.shape
rgb_image = np.zeros((height, width, 3), dtype=np.uint8)

# Color pixels based on likelihood
for i in range(height):
    for j in range(width):
        if val1[i, j] > val2[i, j]:
            rgb_image[i, j] = (np.array([0, 0, 255])*(val1[i,j]/255) + ((255-val1[i,j])/255) *np.array([255,255,255])) # Blue
        else:
            rgb_image[i, j] = (np.array([255, 0, 0])*(val2[i,j]/255) + ((255-val2[i,j])/255) *np.array([255,255,255])) # Blue

# Display the result
plt.imshow(rgb_image)

plt.title('Likelihood Visualization')
plt.show()
```
