```python
#This notebook contains a method by which we can find the number of XRF lines from a given data
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import pandas as pd
```

```python
#Data (Source = https://data.mendeley.com/datasets/nkpmdtdkfw/1)

energy_df = pd.read_csv("Energy.csv")
energy = energy_df.values
print(energy)
```

```python
#Now we visualize the data
plt.title("Base Data")
plt.plot(energy)
plt.show()
```

```python
#Detecting peaks using scipy
x = energy[:,0]
y = energy[:,1]

peak, _ = find_peaks(y,height = 5000,distance=500)

print(peak)
```

```python
plt.plot(x, y, label='XRF spectrum')
plt.plot(x[peak], y[peak], 'x', label='Detected peaks')
plt.title('Detected Peaks')
plt.xlabel('Energy')
plt.ylabel('Intensity')
plt.legend()
plt.show()

print("Energies at which peaks have been detected")
print(peak)
```
